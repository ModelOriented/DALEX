```python
from evaluation.config import DATA_DIR
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
ESTIMATES_DIR = DATA_DIR / "estimates" 
```

```python
def combine(dataset: str, dir: str):
    dfs = []
    for file in sorted(os.listdir(DATA_DIR / dir / "single")):
        if file.endswith(".parquet") and file.startswith(dataset):
            df = pd.read_parquet(DATA_DIR / dir / "single" / file)
            if len(df) != 0:
                print(file)
                # display(df)
                dfs.append(df)
    pd.concat(dfs).reset_index(drop=True).to_parquet(
        DATA_DIR / dir / "combined" / f"{dataset}.parquet"
    )
```

```python
combine("cancer", dir="estimates")
```

```python
combine("cancer", dir="times")
```

```python
combine("housing", dir="estimates")
```

```python
combine("housing", dir="times")
```
