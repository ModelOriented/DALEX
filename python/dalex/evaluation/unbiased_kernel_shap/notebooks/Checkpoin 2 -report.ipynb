## 1. Topic

The goal of this project is to extend Dalex[^1] by Regression Shap[^2]. 

## 2. Work

We forked Dalex repository and implemented new class called Unbiased Kernel Shap which is expected to calculate estimated Shapley Values at it was in the paper[^2]. We wrote unit tests for this class and its method and successfuly installed library via pip. Our experiments were performed on Titanic[^3] dataset. We have trained three different binary classification models to compare results: Random Forrest, Support Vector Machine and Logistic Regression. Tests checked influence of different number of samples on performace. We compared results  with regular Kernel Shap method form library shap[^4]. We calculated errors from both methods compering to exact shapley values. The results are presented in the next paragraph.

## 3. Results
